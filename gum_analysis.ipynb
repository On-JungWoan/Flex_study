```python
import pandas as pd

file_path = 'c:/users/samsung/musicData2.csv'
music_data = pd.read_csv(file_path); music_data
```

```python
music_data.info()
```

```python
music_data['Music Count'].max()
```

```python
bins = [0,10,20,30]
lab = ['0~10','10~20','20~30']
music_data['Count_Info'] = pd.cut(music_data['Music Count'],bins,labels=lab)
```

```python
DC_Group = music_data.groupby(['Distribution','Count_Info']).mean().round(2)
DC_Group
```

```python
import matplotlib.pyplot as plt

plt.plot(DC_Group)
plt.show()
```
