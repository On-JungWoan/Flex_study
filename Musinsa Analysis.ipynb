```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
title = []
price = []

for page in range(10):
    url = 'https://search.musinsa.com/ranking/best?period=now&age=ALL&mainCategory=&subCategory=&leafCategory=&price=&golf=false&newProduct=false&exclusive=false&discount=false&soldOut=false&page=' + str(page) + '&viewType=small&device=&priceMin=&priceMax='
    html = urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')

    
    item_title = soup.find_all(class_='item_title')
    item_price = soup.find_all(class_='price')
    
    #자꾸 item_title의 개수가 1~개 더 많은 문제 해결
    if len(item_title) != 90:
        del item_title[90-len(item_title):]
        
    for i in range(len(item_title)):
        item_title[i] = item_title[i].text.strip('\n')
        title.append(item_title[i])

    for i in range(len(item_price)):
        item_price[i] = item_price[i].text.replace(' ','').strip('\n')
        item_price[i] = int(item_price[i].split('\n')[-1].replace(',','').strip('원'))
        price.append(item_price[i])
        
```

```python
import pandas as pd

data = {'브랜드명':title,'가격':price}
Df = pd.DataFrame(data)
Df
```

```python
Df_price = Df.groupby('브랜드명').mean().astype('int')
Df_price.sort_values(by='가격',ascending=False).head(10)
```

```python
Df_tmp = Df.groupby('브랜드명').count()
Df_tmp.sort_values(by='가격',ascending=False).head(10)
```
